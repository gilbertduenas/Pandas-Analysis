## Setup and Initialization

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# path to csv
csv_path = "Resources/purchase_data.csv"

# read csv and store in a dataframe
purchase_data = pd.read_csv(csv_path) 
purchase_data.head()
```

## Player Count
* Display the total number of players

```python
# count screennames and print the result
player_count = purchase_data['SN'].count()
player_count
```

## Purchasing Analysis (Total)
* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# get a list of unique items and count the items in the list
purchase_data_unique_items = purchase_data['Item Name'].value_counts()
len(set(purchase_data_unique_items))
```

```python
# describe the price data and store the table in a dataframe.
# change the data types and formatting
purchase_data_summary = purchase_data.describe()
purchase_data_summary = purchase_data_summary.astype({'Purchase ID': int, 'Age': int, 'Item ID': int, 'Price': float})
pd.options.display.float_format = '{:,.2f}'.format
purchase_data_summary
```

```python
# find total revenue
purchase_data['Price'].sum()
```

## Gender Demographics
* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
# get a list of totals for different genders
purchase_data_gender = purchase_data['Gender'].value_counts()

# get the male count and divide by the total count.
male_count = purchase_data_gender['Male']
male_percent = male_count / player_count * 100

# get the female count and divide by the total count.
female_count = purchase_data_gender['Female']
female_percent = female_count / player_count * 100

# get the other count and divide by the total count.
other_count = purchase_data_gender['Other / Non-Disclosed']
other_percent = other_count / player_count * 100

# create a dataframe with percents and counts
gender_demographics = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
    'Percent': [male_percent, female_percent, other_percent],
    'Count': [male_count, female_count, other_count]})
gender_demographics
```

## Purchasing Analysis (Gender)
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# describe the price data for gender groups and store the table in a dataframe.
# change the data type and formatting
purchase_data_gender_summary = purchase_data.groupby('Gender')[['Price']].describe()
purchase_data_gender_summary = purchase_data_gender_summary['Price'].astype({'count': int})
pd.options.display.float_format = '{:,.2f}'.format
purchase_data_gender_summary
```

```python
# calculate total purchases by gender group
purchase_data_gender_price_summary = purchase_data.groupby('Gender')[['Price']].sum()
purchase_data_gender_price_summary
```

## Age Demographics
* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
# create bins and names
bins = [0, 9, 19, 29, 99]
bin_names = ["9 and younger", "10 to 19", "20 to 29", "30 and older"]

# create a list of ages and apply the bins and names
purchase_data_age_summary = purchase_data[['Age', 'Price']]
purchase_data_age_summary["Age Groups"] = pd.cut(purchase_data_age_summary['Age'], bins, labels = bin_names)
purchase_data_age_summary_unique = purchase_data_age_summary['Age Groups'].value_counts()

# create a grouped summary of the age data
purchase_data_age_summary_group = purchase_data_age_summary.groupby('Age Groups')[['Age']].describe()
purchase_data_age_summary_group = purchase_data_age_summary_group['Age'].astype({'count': int, 'min': int, 'max': int})
pd.options.display.float_format = '{:,.2f}'.format
purchase_data_age_summary_group
```

## Purchasing Analysis (Age)
* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# describe the price data for age groups and store the table in a dataframe.
# change the data types and formatting
purchase_data_price_summary_group = purchase_data_age_summary.groupby('Age Groups')[['Price']].describe()
purchase_data_price_summary_group = purchase_data_price_summary_group['Price'].astype({'count': int})
pd.options.display.float_format = '{:,.2f}'.format
purchase_data_price_summary_group
```

```python
# show price data for each age
purchase_data_each_age_summary = purchase_data.groupby('Age')[['Price']].describe()
purchase_data_each_age_summary = purchase_data_each_age_summary['Price'].astype({'count': int})
pd.options.display.float_format = '{:,.2f}'.format
purchase_data_each_age_summary
```

## Top Spenders
* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
# sort purchase data by highest amount spent by a user
purchase_data[['SN', 'Price']].sort_values('Price', ascending=False).nlargest(20, 'Price')
```

## Most Popular Items
* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Display a preview of the summary data frame

```python
# create a dataframe with Item ID, Item Name, and Price
purchase_data_ID_Name_Price = purchase_data[['Item ID', 'Item Name', 'Price']]
purchase_data_ID_Name_Price.groupby(['Item ID', 'Item Name']).size().reset_index(name='Total Count').sort_values('Total Count', ascending = False)
```

```python
# group by item name and sort by price
purchase_data_ID_Name_Price.groupby(['Item Name']).apply(lambda x: x.sort_values(['Price'], ascending = False))
```

```python
# group by item id and sort by total price
purchase_data_ID_Name_Price.groupby(['Item ID', 'Item Name']).sum().sort_values('Price')
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the data frame

```python
# group by item id and sort by total price descending
purchase_data_ID_Name_Price.groupby(['Item ID', 'Item Name']).sum().sort_values('Price', ascending = False)
```
